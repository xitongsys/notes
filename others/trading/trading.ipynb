# Trading Notes



## [成交量](https://zhuanlan.zhihu.com/p/620792554)

成交量是用来改变趋势的。比如，无论是股票还是期货，在低位横盘震荡，如果成交没有放量的话，一般情况下，趋势是不会改变的，而是继续震荡。然后，突然某一天，低位放量上涨，出现一根标志性 K 线，这个成交量放量往往是用来改变趋势的，打破原来一直震荡的趋势，所以这个时候股票或者期货都是可以在标志性 K 线出现之后买入的。

相反，无论是股票还是期货，在高位横盘震荡，如果没有放量的话，可能还是在高位横盘震荡，突然有一天，高位放量向下突破了，出现一根标志性 K 线了，这个成交放量也是用来改变趋势的，打破原来的震荡去世，开始向下突破，这个时候股票就需要卖出了，期货可以做空了。

所以，这个就是它那句话里面最简单的道理，低位放量买入，高位放量卖出。低买高卖都知道这个道理，但是它强调的是放量，主要是对成交量的理解和运用，成交量是用来改变趋势的。这种低位或者高位出现标志性 K 线，往往就是反转起爆点。

再举个例子，无论是股票还是期货，低位起来之后，开始上涨，在遇到阻力位之前，如果突然放量，这是不对的，不放量才是对的，如果放量的话，可能就要回调了，不放量的话，原来上涨的趋势会继续延续，盘面小碎步低成交量或者温和成交量往上继续推。

但是最终一定会放量的，一般都是到最后突然大爆发的时候，开始放量暴涨，这个时候就是上涨末期的逼空或者疯狂派发行情，所以行情的斜率也变得更加陡峭。所以，在上涨趋势中，成交量保持平稳，盘面价格稳步推进，这个是正常的，股票和期货都可以继续持有，一旦不正常放量了，就需要注意风险了。

有一种情况下，放量是正常的。比如，在上涨趋势中，突然盘面走到了阻力位附近，这个位置，股票的话有很多人被套，期货的话有很多人背靠阻力位做空。如果盘面在这个位置突破，但是并没有放量，往往是假突破。相反，如果在这个位置放量，说明股票的套牢盘解套离场了，依然没有把价格打下来，原趋势继续，或者期货空头在这个位置砍仓认输了甚至反手了，趋势容易继续。

这些基本上都是纯技术范畴了，核心就是一句话：成交量是用来改变趋势的。高低位成交量突然爆量，往往是出现标志性 K 线，行情启动；趋势中成交量应该平稳，一旦放量就需要注意资金出货；阻力位或支撑位的突破时成交量应该放量，一旦缩量往往是假突破。这些原则股票和期货基本上都适用。

这种纯技术分析可以用于任何有价格的资产交易过程中，但是股票和期货还是不太一样的。比如，早期买股票的时候，炒概念比较多，追涨比较多，然后追高了，被套了，就容易割。现在，不买任何概念股，有时候也会追涨，但是股票不同，追涨的结果不一样。

股票每天波动很正常的，有的股票跌下去了，还会涨回来，甚至创新高，而有的股票跌下去了，不仅回不来了，反而还会跌的更低。所以，买高了不一定是风险，而买错股票才是风险。国内的很过概念股，很容易让人买高了，但是却是买错了，是那种一辈子也回不去了的股票。

炒股的话，需求才是硬道理，人会产生需求，看一下人口结构，儿童、青年、老年人的结构比例以及未来 5-10 年的趋势，找到人口趋势，就找到了那些人的刚需，然后找到对应的品种买入，即便当下买高了，跌了下去，后面还会涨上来的，甚至创新高，翻几倍。

所以，我的理解是，做股票的话，买高了不是风险，买错了才是风险。现在慢慢能理解林园说的怕高的都是苦命人那句话的意思了，本质上就是要买对股票，当下可能是高的，有回调风险，但是未来可能更高。相反，去追那些概念股，当下是高的，未来再也回不去了。

此外，金融学把方差、波动定义为风险，这个是不对的。波动不是风险，也不应该是风险。只有穿越波动，才能最终超越波动，你能穿越多大的波动，你就能拥有多大的收益。一点波动就扛不住了，是不赚不到钱的。

过去我讲过，想明白了，才能赚大钱。一个人的财富不会超过他的格局，你不可能赚到你不相信的那个钱。我个人喜欢在无风险的情况下死拿，拿到自己的目标位，至少敢想，结果就三种：到达自己的目标走人、目标没到但是合约快到期了赚多少都走人、没到目标又没到期但给你扫保本止损了。

中间那些期货价格波动很正常，股票和期货价格每天都是波动的，不要把心思和精力浪费在处理波动上面，波动不是你的风险。想赚大钱，首先格局要大，要敢想；其次敢于穿越波动，股票不用设保本，期货最好设保本，然后就去实现你想要的那个行情。

对大多数人来说，波动是他们的风险，当然，波动也是他们的收益来源。但对于股票来说，频繁的买卖，赚波动的钱，本身不创造价值，就是财富的转移，很难赚大钱，赚大钱还是得跟企业价值一起成长。随着企业 EPS 不断增加，市场对其也越来越乐观，给予了更高的 PE，从而使得股价不断上涨。我们赚的是企业价值增长的钱，不是股价频繁波动买卖的钱。期货也是类似的道理，我们赚的供需矛盾从出现到爆发最后到结束的钱，不是盘面波动频繁买卖的钱。

股票和期货有相似的地方，也有不同的地方。如果单纯从技术分析角度来说，没有任何区别，都是看图操作吗，就是价量分析，价格高低以及量能变化。

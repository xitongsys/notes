1. $ReLU = max(0, x)$ nonlinear activation function. 目前看来，并没有特别的性能优势，但是简单。用的较多

2. $f(x) = \frac{1}{1 + e^{-x}}$ Sigmod activation function

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)

sigmod = 1 / (1 + np.exp(x*-1))
tanh = np.tanh(x)

plt.plot(x, sigmod, label='sigmod')
plt.plot(x, tanh, label='tanh')
plt.legend()
```

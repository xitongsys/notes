## 线性规划处理绝对值

对于绝对值$|x|$，可以做如下变量替换$\frac{|x|+x}{2} = a, \frac{|x|-x}{2} = b$，就有$|x| = a + b, x = a - b$

$a,b$ 要有约束条件 $a>=0, b>=0, a*b = 0$

对于$a*b=0$这个条件可以忽略。原因如下

问题提炼为 

目标函数 $T = min\ f(x,y)$，f为一阶线性函数

约束条件 $x*y=0, x>=0, y>=0$ 与 $x>=0, y>=0$等价

原因是f为线性函数，对于一个线性规划问题，最小值一定在边界上（这个边界包括 $\infty$，当然如果取到无穷，说明最小值点不存在)。

$x*y=0, x>=0, y>=0$ 这个条件是把定义域约束到了**x，y的正半轴上**

$x>=0,y>=0$ 这个条件是**x，y正半轴 + 第一象限**。但是因为目标函数是一阶线性规划，所以第一象限的点就没有用了，所以两个约束是一样的

## 拉格朗日乘子法

等式约束问题

目标函数 $T = min f(x)$, x为向量

约束条件 $g(x) = 0$

直观来想这个问题，考虑目标函数的一系列等高线 $f(x) = a$，如果这个等高线与 $g(x)=0$ 这条线相交或者相切，那么 a 就是满足条件的一个值，但未必是最小值。所有等高线中，最小值一定在某条等高线和 $g(x)=0$ 相切的位置。相切也就是切线共线，即 $\nabla f = \lambda \nabla g$，也就是拉格朗日乘子法的结果

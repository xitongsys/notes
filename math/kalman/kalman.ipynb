# Kalman Filter

网上很多关于卡尔曼滤波的讲解，但是都没有讲清楚其最核心的原理。明白了其核心原理，kalman filter是一个很简单的方法。其核心就是 **找到一个相加权重（kalman 增益），使得 估计值+新测量 值的估计方差最小（如果是多个估计量，就是协方差矩阵trace最小）**

考虑最简单的连续测量一个信号，比如温度

1. 每次测量都会有一定的测量误差（方差），设为$R$，每次的测量值为$z_i$

2. 在每次估计的时候，我们根据前面的测量值和最新的测量值，可以得到一个最优估计值，记为$x_i$，这个值也有一个估计误差(方差)，记为$P_i$

问题就是根据前面的最优估计值$x_{i-1}$ 和最新的测量值 $z_i$ 得到当前的最有估计值 $x_i$。最简单的想法就是找个权重组合，将这两个数加起来，即

$x_i = (1-K)x_{i-1} + K z_i$

而标准就是使得最终估计的方差最小，也就是找到 K 使得$var(x_i)$ 最小。而

$var(x_i) = (1-K)^2 var(x_{i-1}) + K^2 var(z_i) => P_i = (1-K)^2 P_{i-1} + K R$ 

对 K 求导就得到最优的 $K = \frac{P_{i-1}}{P_{i-1} + R}$ 这就是kalman 增益


这是最简单的情形，实际求下一个估计值的时候，可能会有其他操作（比如估计飞行器位置，不是简单的将上个位置作为最新的，还要加上 速度x时间。还有一些有外在其他的控制量，比如飞行器有个固定的加速度）。下面推导一般情形

$y_i = F_i x_{i-1} + G_i u_i$

其中$y_i$ 是最新估计值，是个列向量，$x_{i-1}$ 为前一次最优估计值，$F_i,G_i$ 为矩阵，$u_i$ 为控制向量，$w_i$为过程噪声。设$y_i$的协方差矩阵为$P_i$，有

$P_i = F_i P_{i-1} F_i^T$

我们的最优估计值为 $x_i = (I-K)y_i + K H z_i$，其中$z_i$是最新测量向量，其观测协方差矩阵为$R_i$，H 为观测转换矩阵，K为kalman 增益矩阵

现在的目标是使得$Q_i = var(x_i)$的trace最小

$Q_i = (I-K_i) P_i (I-K_i)^T + K_i H_i R_i H_i^T K_i^T = (I-K_i) F_i P_{i-1} F_i^T (I-K_i)^T + K_i H_i R_i H_i^T K_i^T$







为书写简单，先把下标省略

$Q = -K F P F^T - F P F^T K^T + KFPF^TK^T + KHRH^TK^T$

**trace求导公式**

$\frac{d\ tr(AB)}{dA} = B^T$

$\frac{d\ tr(ABA^T)}{dA} = 2AB$

因此我们把$tr(Q)$ 对 $K$求导有 

$\frac{d\ tr(Q)}{d K} = -2FPF^T + 2KHPF^T + 2KHRH^T = 0$ 就求得

$K = FPF^T (FPF^T + HRH^T)^{-1}$，这就是一般形式的kalman增益

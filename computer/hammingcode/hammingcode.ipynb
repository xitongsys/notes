# Hamming Code

这个视频讲很清楚 [【硬件科普】ECC内存是如何发现错误并纠正的？ 硬件茶谈](https://www.youtube.com/watch?v=zzeuOecdgAI)，但是从二维图像出发，不如直接用二进制位表示来解释，更简单。见后文

1. Hamming Code 是一种基偶校验码，会在数据中多加一个bit，如果之前数据有奇数个1,那么这一位就设成1,否则就是0。因此，增加校验位之后，数据段里1的个数永远是偶数

2. 解决的问题是，在一个长的0,1数组里面，尽可能少的校验位，来修正错误。对于hamming code来说，1位翻转，是可以修复的，2位翻转，可以识别，3位翻转，就验证不出了

3. 考虑一个16个位置0-1数组，可以用4个二进制bit表示 `0000` 直到 `1111`。将二进制位置表示的4个位，每个位为1的所有位置，增加一个校验码。比如，第1位为1的有 0001,0011,0101,0101,0111,1001,1011,1111，然后0001就作为校验码。同理，第2位为1的，0010就作为校验码。这样需要4个校验码，同时0000位置也是个校验码，作为全局校验，一共需要5个校验码

4. 这样我们取检验4个校验码所对应位置中1的个数，只要是奇数，那么一定是这个位置为1的那些位置中有错的。如果有一个错误，那么全局校验码为1，设错误位为 1111，依次检验各个位置，比如第1个位置为1的个数是偶数，那么说明第1个位置为1的没有问题，那么错误位变为 1110，以此类推，就可以找到错误位置了。

5. 每个位置为1的含义，比如第2位为1的有，0010,0011,0110,0111,1010,1011,1110,1111，也就是按照2的长度将数组分组，隔1组取一个，得到的结果。从二维来看，就是宽度为1,2,4,8,...的行和列

6. 实际操作不用每个区域去统计，而是将数组中所有为1的位置取异或操作，如果为0,则是正常，如果不是，则有位置出错了，而异或的结果就是出错的位置。因为如果不错，所有位置应该都是偶数个，所以应该为零。如果有一个位置0->1,那么就会增加一个这个位置，根0异或，结果就是这个位置

# My Puzzles 2023

## 毒苹果

一排20个苹果，最后一个是毒苹果。4个人围成一个圈，轮流从前往后取苹果，一次只能拿1个或2个，取到最后一个毒苹果就被毒死。每个人都想尽可能多取到苹果，同时又不被毒死。每个人都非常聪明，按照最优的方法取。请问最后哪个人会被毒死？


### solve

1. 每个人只要保证不被毒死，优先拿两个苹果
2. 从小开始考虑，比如1个苹果，那么第一个人必死，2个平苹果第2个人必死等。下面的表表示x个苹果，对应哪个人必死

```

第1个人 | 1       8      15 

第2个人 | 2 3    9 10   16 17

第3个人 | 4 5    11 12  18 19

第4个人 | 6 7    13 14  20

```

比如有4个苹果，

如果第1个人拿1个，当前还剩下3个苹果给第2个人，这时候把第2个人看作第1个人<br>
从表中可看到，3个苹果，第2个人是必死的，但是此时是把第2个人看作第1个，所以是第3个人必死

如果第1个人拿2个，此时还剩下2个苹果给第2个人，这时候把第2个人看作第1个人<br>
从表中可看到，2个苹果，第2个人是必死的，但是此时是把第2个人看作第1个，所以是第3个人必死

无论拿1个还是2个，都是第3个人必死，而第1个人优先拿2个苹果，所以当4个苹果，必死的就是第3个人

以此类推，就得到了上表。所以第4个人必死

所以有 n 个人，m 个苹果，也是类似的

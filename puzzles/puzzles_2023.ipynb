## 波动率 问题一

有两个数组,长度分别为$n_1$,$n_2$，即

$[x_1] = [x_{1_1},x_{1_2},...x_{1n_1}]$, $[x_2] = [x_{2_1},x_{2_2},...x_{2_{n_1}}]$

其variance分别为 $v_1 = \overline{(x_1 - \overline{x_1})^2}$, $v_2 = \overline{(x_2 - \overline{x_2})^2}$

两个连接成一个数组，长度为$n=n_1 + n_2$

$[x] = [x_{1_1},x_{1_2},...x_{1n_1}, x_{2_1},x_{2_2},...x_{2_{n_1}}]$

其variance为 $v = \overline{(x - \overline{x})^2}$

问：

$v$ 跟 $v_1,v_2$的关系，如果假定$v_1<=v_2$，那么$v$是否一定在$[v_1,v_2]$之间，跟两个的平均值又是什么关系，即$\frac{n_1 v_1 + n_2 v_2}{n}$ ？


### 解答：

最简单的case，左右两边都是两条常数直线，但是数值不一样，那么$v_1 = v_2 = 0, v > 0$，显然可以大于$v_2$

考虑下面的式子

\begin{equation}
s = \frac{1}{n}\sum_{i=1}^{n}(x_i - a)^2
\end{equation}

是个凸函数，有最小值，我们对a求导令等式为0，得到

\begin{equation}
\sum_{i=1}^{n}(x_i - a) = 0
\end{equation}

即 $a = \frac{\sum_{i=1}{n}x_i}{n} = \overline{x}$

即当a为数组平均值时候是最小的

平均的variance为

$v_{avg} = \frac{v_1 n_1 + v_2 n_2}{n} = \frac{n_1}{n} \overline{(x_1 - \overline{x_1})^2} + \frac{n_2}{n} \overline{(x_2 - \overline{x_2})^2}$

 而真实的variance为

$v_{real} = \frac{v_1 n_1 + v_2 n_2}{n} = \frac{n_1}{n} \overline{(x_1 - \overline{x})^2} + \frac{n_2}{n} \overline{(x_2 - \overline{x})^2}$

跟据上面得到的结论就可以得出，$v_{real} >= v_{avg}$ 

所以关系为

$v_1 <= v_{avg} <= v_2$

$v_1 <= v_{avg} <= v_{real}$

## 波动率 问题二

有两个数组,长度分别为$n_1$,$n_2$，即

$[x_1] = [x_{1_1},x_{1_2},...x_{1n_1}]$, $[x_2] = [x_{2_1},x_{2_2},...x_{2_{n_1}}]$

令有两个数组，度分别为$n_1$,$n_2$，即

$[y_1] = [y_{1_1},y_{1_2},...y_{1n_1}]$, $[y_2] = [y_{2_1},y_{2_2},...y_{2_{n_1}}]$


他们的covariance分别为

$c_1 = cov(x_1, y_1) = \overline{(x_1 - \overline{x_1})(y_1 - \overline{y_1})}$

$c_2 = cov(x_2, y_2) = \overline{(x_2 - \overline{x_2})(y_2 - \overline{y_2})}$

分别连接成一个数组，长度为$n=n_1 + n_2$

$[x] = [x_{1_1},x_{1_2},...x_{1n_1}, x_{2_1},x_{2_2},...x_{2_{n_1}}]$

$[y] = [y_{1_1},y_{1_2},...y_{1n_1}, y_{2_1},y_{2_2},...y_{2_{n_1}}]$

这两个数组的covariance为

$c = cov(x, y) = \overline{(x - \overline{x})(y - \overline{y})}$

问：

$c$ 跟 $c_1,c_2$的关系，如果假定$c_1<=c_2$，那么$c$是否一定在$[c_1,c_2]$之间，跟两个的平均值又是什么关系，即$\frac{n_1 c_1 + n_2 c_2}{n}$ ？


### 解答

\begin{equation}
\begin{aligned}

\overline{x} = \frac{n_1 \overline{x_1} + n_2 \overline{x_2}}{n} \\

\overline{y} = \frac{n_1 \overline{y_1} + n_2 \overline{y_2}}{n} 

\end{aligned}
\end{equation}


带入covariance公式得到 

\begin{equation}
\begin{aligned}

c_{real} = \overline{xy} - \overline{x}\overline{y} 

= \frac{nn_1\overline{x_1 y_1} + nn_2\overline{x_2 y_2} - (n_1\overline{x_1} + n_2\overline{x_2})(n_1\overline{y_1} + n_2\overline{x_2})}{n^2} \\


c_{avg} = \frac{n_1 c_1 + n_2 c_2}{n} = \frac{nn_1\overline{x_1 y_1} + nn_2\overline{x_2 y_2} - nn_1\overline{x_1}\overline{y_1} - nn_2\overline{x_2}\overline{y_2}}{n^2}

\end{aligned}
\end{equation}

整理变换就得到

\begin{equation}
c_{real} = c_{avg} + \frac{n_1 n_2}{n^2}(\overline{x_1} - \overline{x_2})(\overline{y_1} - \overline{y_2})
\end{equation}

因为

$c_1 <= c_{avg} <= c_2$，又上面可见$c_{real}$ 可以任意大小，如果$\overline{x_2} > \overline{x_1}$ 且 $\overline{y_2} > \overline{y_1}$，那么$c_{real} > c_{avg}$，而且可以任意大

同理，如果$\overline{x_2} > \overline{x_1}$ 且 $\overline{y_2} < \overline{y_1}$，那么$c_{real} < c_{avg}$，而且可以任意小

这其实就解释了为什么前面计算总的cov，比分两段计算cov的值都大。

## 核酸
10人混管核酸，如果有1人阳性，则每个人都要再做一次单管核酸。如果社会面感染率超过多少，混管成本将大于全部单管成本？

```
设感染率为p,则 10 人全部为阴性的概率为 (1-p)^10, 有阳性的概率为1 - (1-p)^10
所以平均检测次数为

1*(1-p)^10 + 11*(1-(1-p)^10) >= 10

解得 p = 0.20567
```

## 数组diff

给定一个数组，不断取diff，直到最后剩下一个数字，这个数字是多少？

比如[1,2,3,4]->[1,1,1]->[0,0]->[0]

### solve

```
*  * *   *
\  /  \ /
  *    *
   \  /
     *
```

可以看作一棵倒立的树。向右走乘以-1，向左走乘以+1。每个元素最终在结果中出现的次数就是从上往下走，不同路径的数目。而每条路径上向右走的次数为奇数的，符号就是-1。注意到，每个元素，无论按照那条路径走，其向右走和向左走的数目是一样的，也就是每条路径的符号都是一样的。因此，如果第一个元素是+1，第二个元素就是-1,后面依次交替

所以关键就是求得每个元素路径的数目。每个元素的数目就等于下层两个节点的数目之和。因此其数目就是杨辉三角

```
0          1
1         1 1
2        1 2 1
3       1 3 3 1
          ....
n       C(n-1,i)
```

设数组下标从0开始的n个元素的数组为$[a_0,a_1,...a_{n-1}]$

杨辉三角的层数也是从0层开始，因此第n-1层就有n个元素

所以n个元素数组的最终结果 $D(n) = \sum_{i=0}^{n-1}C_{n-1}^i(-1)^{n-1-i}a_i$

````python
import numpy as np
import math,functools

rs = np.array([1,5,2,5])

# solve1
def fn(n):
    res = 0
    for i in range(n):
        res += math.comb(n-1, i)*((-1)**(n-1-i))*rs[i]
    return res

ans1 = fn(len(rs))

# check
ans = rs
for i in range(len(rs)-1):
    ans = np.diff(ans)
ans = ans[0]
    
print(f"solve1={ans1}, real={ans}")
````

一个有趣的现象，我们一步一步diff，每次都将新的图做出来，会发现下面的现象，最终图像会变成一个一个的波浪
原因如下：

考虑数组n个元素的数组$[a_0,a_1,...a_{n-1}]$，如果我们diff m次，得到的数组为$[b_0,b_1,...,b_{n-1-m}]$

其中 (从上面倒树可以看出)

$b_0$ 就是 $[a_0,..,a_{m}]$ diff m次的结果，

$b_1$ 就是 $[a_1,..,a_{m+1}]$ diff m次的结果，

... ...

一个m个元素的数组最终diff到只有一个元素，从上面的结果得到

$D(m) = \sum_{i=0}^{m-1}C_{m-1}^i(-1)^{m-1-i}a_i$

而这个式子的含义就是一个正负交替的 m重二项分布与数组的卷积！！！

不断的diff，就是不断的卷积过程，也就得到了下面的现象，出现的波浪，正是二项分布的形状，随着diff次数增加，波浪会越来越大（类似于一个信号调幅的过程）

**对于金融产品的价格，diff 一次是其return，往往用作动量因子，会有提前效应。但是如果继续diff，也会落入下面的现象，其实就没用了**

````python
import numpy as np
import math,functools
import matplotlib.pyplot as plt

rs = np.random.random(400)

for i in range(len(rs)):
    rs = np.diff(rs)
    if i % 100 == 0:
        plt.plot(rs)
        plt.show()
````

### solve

1. 每个人只要保证不被毒死，优先拿两个苹果
2. 从小开始考虑，比如1个苹果，那么第一个人必死，2个平苹果第2个人必死等。下面的表表示x个苹果，对应哪个人必死

```

第1个人 | 1       8      15 

第2个人 | 2 3    9 10   16 17

第3个人 | 4 5    11 12  18 19

```

比如有4个苹果，

如果第1个人拿1个，当前还剩下3个苹果给第2个人，这时候把第2个人看作第1个人<br>

## 打水问题

公司有个水桶，容量为20L，每个人的杯子容量为6L，每个人打水都是完全随机的，如果桶里的水不够了，当前打水的人需要先把剩余水打完，再换桶继续打，直到自己水杯满了。如果打完水，水桶中没有水了，也需要换水。请问，

1. 当你去打水，需要换水的概率？

2. 如果水桶容量 V 和杯子容量 A 不一定是整数，有什么普遍的结论？

3. 如果每个人杯子不一样，比如三个人，杯子容量分别为A，B，C，每个人去打水的频率是一样的，那么每个人取打水，需要换水的概率是多少？

### solve

1. 

20和6的最小公倍数是60，所以3桶水，10次打水，再往后就是循环的了。10次打水中，只有3次是需要换水的（因为换了3桶水），所以换水概率就是 3/10 = 0.3


2. 

设桶容量是 V，杯子容量为 A，其实就是找到 m * V = n * A，m,n为整数。那么换水的概率就是 

$ p = \frac{m}{n} = \frac{A}{V} $

如果A，B都是有理数，m,n一定存在的。如果其中有无理数，这个式子可能不存在了。对于无理数，可以用有理数逼近，比如都只取小数点后10000位，那么都是有理数了，上面的结论依然成立，取极限，结论依然成立。

因此需要换水的概率就是 $ p = min(1, \frac{杯子容量}{水桶容量}) $

3. 

考虑无限次打水的样本空间，因为每个人打水频率是一样的，所以每个人打水的次数的期望是一样的，设为n次，那么总共打水次数为3n次。一共打水量为 $n\times(A+B+C)$，需要 $m = \frac{n\times(A+B+C)}{V}$ 桶水，也就是换水m次。

对于这三个人来讲，其打水概率的比值就是$A:B:C$，a的换水次数的期望为 $n_a = m \frac{A}{A+B+C}$，所以a换水的概率为$p_a = \frac{n_a}{n} = \frac{A}{V}$

如果有N个人，杯子容量为$[A_1,A_2,...,A_N]$，水桶容量为V，每个人每次打水需要换桶的概率为 $p_i = \frac{A_i}{V}$

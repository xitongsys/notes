#### P6. 
   
给定一个独立同分布序列a=$[a_1, a_2...,a_H]$，其中$H$远大于$R$

对其rolling(R).sum()得到新的序列为$A$（即$A_i = a_{i-R+1} + a_{i-R+2} + ... + a_i$），显然$A$的长度为$H-R+1$

将序列$A$序列向前shift(-R) 得到新的序列 $B$（即$B_i = A_{i+R}$），显然$B$的长度为$H-2R+1$

删除$A$序列最后$R$个元素(因为这些元素没有对应的$B$中的元素)，将$A,B$对齐，此时$A,B$长度为$L=H-2R+1$。

$A,B$序列的均值为$\mu_A=\frac{1}{L}\Sigma A_i，\mu_B=\frac{1}{L}\Sigma B_i$

直接计算$A,B$两个序列的协方差，即$cov = \frac{1}{L}\Sigma_{i=1}^L (A_i - u_A)(B_i - u_B)$，证明 $cov$的期望 $E(cov) < 0$，尽可能写出$E(cov)$的表达式
